```python
import pandas as pd
import numpy as np
from IPython.display import Image
import os
import glob
```

# Getting images and their labels

```python
# Base directory where the images are stored
base_dir = "drive/MyDrive/bird recognition/CUB_200_2011/images"
```

```python
# Get image paths and labels
image_paths = glob.glob(os.path.join(base_dir, "*", "*.jpg"))

# Extract class names from the folder structure
labels = [os.path.basename(os.path.dirname(path)) for path in image_paths]

labels = [folder_name.split('.', 1)[1] for folder_name in labels]

labels
```

```python
len(labels)
```

```python
# Find the unique label values
unique_breeds = np.unique(labels)
len(unique_breeds)
```

```python
unique_breeds
```

```python
# Turn a single label into an array of booleans
print(labels[0])
labels[0] == unique_breeds
```

```python
# Turn every label into a boolean array
boolean_labels = [label == unique_breeds for label in labels]
boolean_labels[:2]
```

# Creating Validation Set

```python
# Setup X & y variables
X = image_paths
y = boolean_labels
```

```python
NUM_IMAGES = 1000 #@param {type:"slider", min:1000, max:10000, step:1000}
```

```python
# split our data into train and validation sets
from sklearn.model_selection import train_test_split

# Split them into training and validation of total size NUM_IMAGES
X_train, X_val, y_train, y_val = train_test_split(X[:NUM_IMAGES],
                                                  y[:NUM_IMAGES],
                                                  test_size=0.2,
                                                  random_state=42)

len(X_train), len(y_train), len(X_val), len(y_val)
```
